```python
import numpy as np#linear algebra
import pandas as pd#data processing ,csv file
import matplotlib.pyplot as plt#data visualization
import seaborn as sns#data visualization
df=pd.read_csv('/content/drive/MyDrive/heart.csv')
df.head()
```

```python
df.info()
```

```python
#check for NAN value
df.isnull().sum()
```

```python
df.shape
df.describe()
```

```python
#target distribution
df.target.value_counts()
```

```python
#show correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
#plt target dist
#sns.countplot(df.target,hhue='sex',palette=['green','red']
#sns.countplot(df.target,hue=df,palette=['#1CA53B','red'])
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
#seaborn.pairplot(df,hue='size')
plt.xticks([0,1],labels=['Not disease','Disease'])
plt.title('Target Distribution')
plt.show()
```

```python
#plt target dist
#sns.countplot(df.target,hhue='sex',palette=['green','red']
#sns.countplot(df.target,hue=df,palette=['#1CA53B','red'])
sns.countplot(x=df.target,hue='target',data=df,palette=['#1CA53B','red'])
#seaborn.pairplot(df,hue='size')
plt.xticks([0,1],labels=['Not disease','Disease'])
plt.title('Target Distribution')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend(["Haven't Disease","Have Disease"])
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#sns.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.displot(df.age[df['target']==0],bins=30,color='green',label='Not Disease')
sns.displot(df.age[df['target']==1],bins=30,color='red',label='Disease')
plt.legend()
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
